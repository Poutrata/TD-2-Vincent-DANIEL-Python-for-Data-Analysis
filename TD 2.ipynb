```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
country = pd.read_csv("Country.csv",sep=",")
```

```python
country
```

What is the average year when industrial data was collected ?

```python
country.LatestIndustrialData.agg("mean")
```

In what country do we have the oldest data and what year was it last refreshed?

```python
country.groupby('LatestPopulationCensus').ShortName.sum()[:1]
```

What country are counted as in South Asia ?

```python
country[(country.Region=="South Asia")].groupby('ShortName').Region.sum()
```

Graph of date of aquisition of trade and water withrawl data

```python
country.groupby('LatestIndustrialData').sum().plot()
```

```python
country.groupby('LatestTradeData').sum().plot()
```

These graphs look really different, while they really shouldn't as they are depecting quite similar values, which is really weird.

```python
table = pd.pivot_table(country, values='LatestWaterWithdrawalData', index=['ShortName'], columns=['Region'], aggfunc=np.sum)
```
